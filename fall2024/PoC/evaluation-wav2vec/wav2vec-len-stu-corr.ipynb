# Correlation Between Length of Audio & Stuttering Count [Wav2vec]

```python
import pandas as pd
import numpy as np
import sys
import os
import matplotlib.pyplot as plt
from tqdm import tqdm
import re
```

## Data Ingestion

```python
cols = list(pd.read_csv("net_aigenerated_data_wav2vec_performance_stu.csv", nrows=1))
print(cols)

net_aigenerated_data_wav2vec = pd.read_csv('/projectnb/ds549/projects/AImpower/evaluation-wav2vec/net_aigenerated_data_wav2vec_performance_stu.csv', delimiter=',', usecols =[i for i in cols if "Unnamed:" not in i])
```

```python
net_aigenerated_data_wav2vec['Duration'] = net_aigenerated_data_wav2vec['End_time'] - net_aigenerated_data_wav2vec['Start_time']
```

## Visualization of Correlation

```python
plt.figure(figsize=(6, 4))
plt.scatter(
    net_aigenerated_data_wav2vec["Duration"], 
    net_aigenerated_data_wav2vec["Stutterance_Count"], 
    alpha=0.7  # Handle overlapping points
)

plt.title("Audio Duration to Stuttering Count Relationship", fontsize=16)
plt.xlabel("Audio Duration", fontsize=14)
plt.ylabel("Stuttering Count", fontsize=14)
plt.grid(True)
plt.show()
```
