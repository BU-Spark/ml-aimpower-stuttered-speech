## Comparison Analysis

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
model = ["Whisper Large", "Whisper Tiny", "WeNet", "Wav2Vec", "Google Cloud", "Azure"]
mean_wer = [0.317117, 0.662262, 0.265648, 0.367528, 0.303750, 0.210741]
std_wer = [0.279339, 0.985791, 0.177740, 0.206969, 0.201062, 0.261580]

# Plotting the data with error bars
plt.figure(figsize=(10, 6))
x_pos = np.arange(len(model)) 

plt.bar(x_pos, mean_wer, yerr=std_wer, capsize=5, alpha=0.75, ecolor='black', label='WER')
plt.xticks(x_pos, model, rotation=45, ha='right', fontsize=12)
plt.xlabel("Model", fontsize=14)
plt.ylabel("WER (Mean  Std Dev)", fontsize=14)
plt.title("WER Comparison Across Models", fontsize=16)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.legend()
plt.show()
```

## Correlations [Stuttering Count / Character vs Performance]

```python
models = ["Whisper Large", "Whisper Tiny", "WeNet", "Wav2Vec", "Google Cloud", "Azure"]
stuperchar_rouge1_rho = [0.16211, 0.25046, 0.10860, 0.03752, -0.09062, -0.25807]
stuperchar_rouge2_rho = [0.13050, 0.28003, 0.05129, 0.03700, -0.11115, -0.30123]
stuperchar_rougel_rho = [0.13177, 0.24338, 0.06382, 0.01482, -0.13236, -0.34863]
stuperchar_wer_rho = [-0.05560, -0.10994, -0.00130, 0.11930, 0.20792, 0.45733]

correlation_data = {
    "Rouge 1 Rho": stuperchar_rouge1_rho,
    "Rouge 2 Rho": stuperchar_rouge2_rho,
    "Rouge L Rho": stuperchar_rougel_rho,
    "WER Rho": stuperchar_wer_rho
}

for metric, values in correlation_data.items():
    plt.figure(figsize=(8, 6))
    plt.bar(models, values, color='skyblue', alpha=0.8)
    plt.axhline(0, color='red', linewidth=1, linestyle='--')  # Reference line at 0
    plt.title(f"{metric} Correlation Coefficients", fontsize=16)
    plt.xlabel("Model", fontsize=14)
    plt.ylabel("Correlation Coefficient (Rho)", fontsize=14)
    plt.xticks(rotation=45, fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
```
